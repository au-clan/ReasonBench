```python
import os
import anthropic
from openai import OpenAI
```

API and `reasoning_effort` chosen randomly just for demonstrating cases. <br>
GPT-4.1 Nano is not a reasoning model (control group)

```python
client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=os.getenv("OPENROUTER_API_KEY"),
)

models = [
    "openai/gpt-4.1-nano", # Test (not a reasoning model)
    "deepseek/deepseek-r1",
    "qwen/qwen3-235b-a22b-thinking-2507",
    "openai/gpt-oss-120b",
    "openai/gpt-5-nano",
    "openai/gpt-5-mini",
    # "openai/gpt-5-mini", #Not wasting tokens on this (if mini&nano are, then so is  this)
    "google/gemini-3-flash-preview",
    "x-ai/grok-4.1-fast",
    "moonshotai/kimi-k2-thinking",
    "z-ai/glm-4.5-air",
    "openai/o4-mini",
    "openai/o3-mini",
    "deepseek/deepseek-v3.2"
]
```

```python
for model in models:
    print(f"Model: {model}")

    completion = client.chat.completions.create(
        model=model,
        messages=[
            {
                "role": "user",
                "content": "When I was 6 years old my sister was half my age. Now I am 70 years old. How old is my sister?"
            }
        ],
        max_completion_tokens=50,
        reasoning_effort= "low"
    )


    print(f"Reasoning tokens: {completion.usage.completion_tokens_details.reasoning_tokens}")
    print("---")
```

## Anthropic
Antrhopic in general doesn't expose number of reasoning tokens. But below it's shown how the response is split and it has a thinking part.

```python
client = anthropic.Anthropic()

models = [
    "claude-haiku-4-5",
    "claude-sonnet-4-6"

]

for model in models:
    response = client.messages.create(
        model="claude-haiku-4-5",
        max_tokens=1200,
        thinking={"type": "enabled", "budget_tokens": 1024},
        messages=[
            {
                "role": "user",
                "content": "When I was 6 years old my sister was half my age. Now I am 70 years old. How old is my sister?",
            }
        ],
)

    print(f"Model: {model}")
    print(f"Thinking characters: {len(response.content[0].thinking)}")
    print(f"Output characters: {len(response.content[1].text)}")
    print("---")
```
